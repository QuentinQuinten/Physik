# A Young Man's Game?
## Are great scientific acheivements (in physics) only made during youth?

```python
import matplotlib.pyplot as plt               # for visuals
from matplotlib.ticker import MultipleLocator
%matplotlib inline
import pandas as pd                           # for data table manipulation
from scipy.stats import lognorm               # for statistics
```

### Important Publications

```python
# import the csv data
data = pd.DataFrame(pd.read_csv("data.csv", sep=",", header=0))
data
```

```python
# get relevant data
pub_age_list = [age for age in data[:]["Author Age"]]
print(pub_age_list)
print("\nmaximum age: " + str(max(pub_age_list)))
print("minimum age: " + str(min(pub_age_list)))
# display the age distribution as a histogram for when papers are published
# plot style
plt.style.use("default")
plt.style.use("seaborn-dark")
plt.style.use("grayscale")
# create figure
fig = plt.figure(figsize=(10,4), dpi=300)
fig.patch.set_facecolor('white')
# create axes with histogram    
ax = plt.Axes(fig=fig, rect=[0,0, 1, 1])

ax.hist(pub_age_list, bins=16, range=None, density=True, alpha=0.2, histtype='bar', 
                       align='mid', orientation='vertical', rwidth=None, 
                       log=False, color="black", label=None, stacked=False)
# fit with lognormal distribution
shape,loc,scale = lognorm.fit(sorted(pub_age_list))
pdf = lognorm.pdf(sorted(pub_age_list), shape, loc, scale)
ax.plot(sorted(pub_age_list), pdf, lw=4)
# set axes specifications for visual clarity
ax.xaxis.set_major_locator(MultipleLocator(10))
ax.xaxis.set_minor_locator(MultipleLocator(1))
ax.tick_params(which='both', width=1.5)
ax.tick_params(which='minor', length=4, color="black")
ax.tick_params(which='major', length=7, color="black", labelsize=16)
ax.set_yticks([])
ax.set_xlim([22, 87])
ax.set_xlabel("Age", size=20)
ax.set_title("Author's Age Distribution for Important Publications in Physics", size=18)
# add axes to figure
fig.add_axes(ax)
# plot the whole thing
plt.savefig("physics_age_distribution", bbox_inches="tight", facecolor="white")
```

### Age Groups

```python
# import the csv data
data = pd.DataFrame(pd.read_csv("data_2.csv", sep=",", header=0))
data
```

```python
age_groups = [g for g in data[:]["Age group +"]]
numbers = [n for n in data[:]["Numbers of observations"]]
# normalize numbers
total = 0
NCONST = 4.3
for n in numbers:
    total += n
for i in range(len(numbers)):
    numbers[i] = numbers[i] / (NCONST * total)
    age_groups[i] += 2.5 # + 2.5 because 5 year intevalls
    
# display the age distribution as a histogram for when papers are published
# plot style
plt.style.use("default")
plt.style.use("seaborn-dark")
plt.style.use("grayscale")
# create figure
fig = plt.figure(figsize=(10,4), dpi=300)
fig.patch.set_facecolor('white')
# create axes with histogram    
ax = plt.Axes(fig=fig, rect=[0,0, 1, 1])
# data
ax.plot(age_groups, numbers, "o", ms=8.5, alpha=1)
# Fitting with curv_fit
from scipy.optimize import curve_fit
import numpy as np
# log-norm function
def log_fit(x, a, mu, sigma):
    return a / x * 1. / (sigma * np.sqrt( 2. * np.pi ) ) * np.exp( -( np.log( x ) - mu )**2 / ( 2. * sigma**2 ) )
# first curve
popt1, pcov = curve_fit(log_fit, age_groups[:4], numbers[:4])
x = np.arange(22,87,0.3)
ax.plot(x, log_fit(x, *popt1), "-", lw=4, color="black", alpha=0.5) ##149349
# second curve
popt2, pcov = curve_fit(log_fit, age_groups[4:], numbers[4:])
ax.plot(x, log_fit(x, *popt2), "--", lw=4, color="black", alpha=0.2)
# total curve (if you are interested)
#popt3, pcov = curve_fit(log_fit, age_groups[:], numbers[:])
#ax.plot(x, log_fit(x, *popt3), lw=4, color="black", ls="--", alpha=0.5)

# add text for clarity
ax.text(33, 0.052, "PhDs & postdocs", size=15)
ax.text(48, 0.028, "Professors & associated professors", size=15)

# set axes specifications for visual clarity
ax.xaxis.set_major_locator(MultipleLocator(10))
ax.xaxis.set_minor_locator(MultipleLocator(1))
ax.tick_params(which='both', width=1.5)
ax.tick_params(which='minor', length=4, color="black")
ax.tick_params(which='major', length=7, color="black", labelsize=16)
ax.set_xlim([22, 87])
ax.set_ylim([0, 0.06])
ax.set_yticks([])
ax.set_xlabel("Age", size=20)
ax.set_title("General Age Distribution in Natural Sciences & Mathematics", size=18)

# add axes to figure
fig.add_axes(ax)
# plot the whole thing
plt.savefig("academic_age_distribution", bbox_inches="tight", facecolor="white")
```
